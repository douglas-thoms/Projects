```r
x = rnorm(100)
x_bar = mean(x)
```

```r
y= 2*x+0.1*rnorm(100)
y_bar= mean(y)
df = data.frame(cbind(x,y))
df
```

```r
library(dplyr)
df <- df %>% 
    mutate(yi_minus_y_bar = y - y_bar,
           xi_minus_x_bar = x - x_bar)
df
```

```r
b = sum((df$xi_minus_x_bar * df$yi_minus_y_bar)) / sum((df$xi_minus_x_bar^2))
```

```r
b
```

```r
a = y_bar - b*x_bar
```

```r
a
```

```r
y_hat = a+b*x
```

```r
y_hat
```

```r
plot(y_hat,y)
```
