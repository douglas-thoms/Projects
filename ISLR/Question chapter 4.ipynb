```r
Power2 <- function(num,a){
    return(num^a)
}
```

```r
Power2(3,8)
```

```r
Power3 <- function(num,a){
    result <- num^a
    return(result)
}
```

```r
x <- 1:10
y <- sapply(x, function(x)Power3(x,3))
plot(x,y)
```

```r
PlotPower <- function(num,a){
    x <- 1:num
    y <- sapply(x, function(x)Power3(x,3))
    plot(x,y)
}
```

```r
PlotPower(10,3)
```

```r
df <- Boston
mediancrim <- median(df$crim)
crim01 <- sapply(df$crim, function(x) {ifelse(x > mediancrim,1,0)})
df$crim01 <- crim01
df <- df[,-1]
df
```

Test_train Split

```r
train_index <- sample(1:nrow(df), 0.8 * nrow(df))
test_index <- setdiff(1:nrow(df), train_index)
```

LDA

```r
glm.prob <- predict(glm.fit,type="response",df[test_index,])
glm.prob
```

```r
glm.pred <- rep(0,nrow(Auto2[test_index,]))
glm.pred[glm.prob>.5] <- "1"
glm.pred
```

```r
lda.fit <- lda(crim01~indus+nox+age+dis+rad+tax+ptratio+black+lstat+medv,data=df,subset=train_index)
lda.fit
```

```r
lda.prob <- predict(lda.fit,df[test_index,])$class
lda.prob
```

```r
lda.pred <- rep(0,nrow(df[test_index,]))
lda.pred[lda.prob == "1"] <- "1"
lda.pred
```
